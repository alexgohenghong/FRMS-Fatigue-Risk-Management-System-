```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# 1. 模拟生成 10,000+ 飞行班表
def generate_aviation_roster(n=10000):
    np.random.seed(42)
    start_date = datetime(2026, 6, 1)
    roster = []

    for i in range(n):
        dep_time = start_date + timedelta(minutes=np.random.randint(0, 43200))
        duty_length = np.clip(np.random.normal(8, 3), 2, 16)
        arr_time = dep_time + timedelta(hours=duty_length)

        roster.append({
            'Flight_ID': f'SQ{np.random.randint(100, 999)}',
            'Dep_Time': dep_time,
            'Arr_Time': arr_time,
            'Duty_Length': round(duty_length, 2)
        })
    return pd.DataFrame(roster)

# 2. 疲劳风险评分算法
def analyze_fatigue(row):
    score = row['Duty_Length'] * 5  # 基础分

    # 逻辑 A: WOCL 生理低谷期判定 (02:00 - 06:00)
    dep_hour = row['Dep_Time'].hour
    if 2 <= dep_hour <= 6:
        score += 30

    # 逻辑 B: EASA FTL 规章检查 (超过13小时报警)
    if row['Duty_Length'] > 13:
        score += 50

    return score

# 3. 执行分析
df = generate_aviation_roster(10000)
df['Fatigue_Score'] = df.apply(analyze_fatigue, axis=1)
df['Is_Red_Zone'] = df['Fatigue_Score'] > 85

# 输出统计总结
print(f"Total Rosters Analyzed: {len(df)}")
print(f"Red Zone Alerts Detected: {df['Is_Red_Zone'].sum()}")
print("\n--- High Risk (Red Zone) Samples ---")
print(df[df['Is_Red_Zone']].head())
```
